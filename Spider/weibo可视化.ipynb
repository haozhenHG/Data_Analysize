```python
import pandas as pd
from pyecharts.globals import CurrentConfig, NotebookType

# 设置 pyecharts 的全局配置，指定图表渲染环境为 Jupyter Lab
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
wb = pd.read_csv('weibodata.csv')
wb.head()
```

```python
# 导入可视化配置项
from pyecharts import options as opts
# 图像类型设置
from pyecharts.charts import Pie
# 导入可视化数据 （随机生成）
from pyecharts.faker import Faker

c = (
    Pie()
    .add(
        "",
        [
            list(z)
            for z in zip(
                # 传入数据内容
                Faker.choose() + Faker.choose() + Faker.choose(),
                Faker.values() + Faker.values() + Faker.values(),
            )
        ],
        center=["40%", "50%"],
    )
    .set_global_opts(
        # 设置可视化标题
        title_opts=opts.TitleOpts(title="Pie-Legend 滚动"),
        legend_opts=opts.LegendOpts(type_="scroll", pos_left="80%", orient="vertical"),
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    # 使用 render 方法将图表渲染到 "pie_scroll_legend.html" 文件中
    # .render("pie_scroll_legend.html")
)
c.load_javascript()
```

```python
c.render_notebook()
```

```python
x = wb['地区'].value_counts().index.tolist()
y = wb['地区'].value_counts().tolist()
```

```python
c = (
    Pie()
    .add(
        "",
        [
            list(z)
            for z in zip(
                # 传入数据内容
                x,
                y,
            )
        ],
        center=["40%", "50%"],
    )
    .set_global_opts(
        # 设置可视化标题
        title_opts=opts.TitleOpts(title="论用户地区分布情况"),
        legend_opts=opts.LegendOpts(type_="scroll", pos_left="80%", orient="vertical"),
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    # 使用 render 方法将图表渲染到 "pie_scroll_legend.html" 文件中
    # .render("pie_scroll_legend.html")
)
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker


c = (
    Bar()
    .add_xaxis(gx)
    .add_yaxis("商家A", gy)
    .add_yaxis("商家B", gy)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Bar-Y 轴 formatter"),
        yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(formatter="{value} /月")),
    )
    # .render("bar_yaxis_formatter.html")
)
c.render_notebook()
```
