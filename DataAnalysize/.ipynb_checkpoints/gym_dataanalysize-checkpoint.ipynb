```python
df_scaled = df
# 筛选出数值型特征的所有列名
numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
# 对数值型特征进行标准化
scaler = StandardScaler()
df_scaled[numerical_features] = scaler.fit_transform(df[numerical_features])
df_scaled.head()
```

```python
# 对分类变量进行独热编码
categorical_features = ['Gender', 'Workout_Type']
df_scaled = pd.get_dummies(df_scaled, columns=categorical_features)
df_scaled.head()
```

```python
df_scaled.iloc[:,-6:] = df_scaled.iloc[:,-6:].astype(int)
df_scaled.head()
```

- 确定聚类数

肘部法则（Elbow Method）
肘部法则是一种用于确定最佳聚类数量的方法，它基于聚类的惯性（inertia），即聚类内样本点到其质心的距离平方和。随着聚类数量的增加，惯性通常会减少，但在某一点之后，增加聚类数量带来的改善会显著减少，这个点就像肘部一样，因此得名肘部法则。

轮廓系数（Silhouette Score）
轮廓系数是一种衡量聚类效果好坏的指标，其值范围从 -1 到 1。轮廓系数综合考虑了聚类的凝聚度（cohesion）和分离度（separation）。较高的轮廓系数表示聚类效果好，即同一聚类内的点紧密相连，而不同聚类之间的点相隔较远。

```python
# 使用肘部法则来确定最佳聚类数
from sklearn.metrics import davies_bouldin_score

inertia = []
silhouette_scores = []
# 初始化戴维斯-邦丁指数列表 
# 指数值越低，表示聚类效果越好。
db_index_list = []
k_range = range(2, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=15).fit(df_scaled)
    inertia.append(kmeans.inertia_) # 聚类的惯性
    silhouette_scores.append(silhouette_score(df_scaled, kmeans.labels_))
    db_index_list.append(davies_bouldin_score(df_scaled, kmeans.labels_))
```

```python
# 设置全局字体大小
plt.rcParams['font.size'] = 14
plt.figure(figsize=(15,20))

plt.subplot(3, 1, 1)
plt.plot(k_range, inertia, marker='o')
plt.xlabel('聚类中心数目')
plt.ylabel('惯性')
plt.title('肘部法则图')

plt.subplot(3, 1, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('聚类中心数目')
plt.ylabel('轮廓系数')
plt.title('轮廓系数图')

# 指数值越低，表示聚类效果越好。
plt.subplot(3, 1, 3)
plt.plot(k_range, db_index_list, marker='o')
plt.xlabel('聚类中心数目')
plt.ylabel('戴维斯-邦丁指数')
plt.title('戴维斯-邦丁指数图')

plt.tight_layout()
plt.show()
```

选择 4 作为聚类数

```python
kmeans = KMeans(n_clusters=4, random_state=15)
kmeans.fit(df_scaled)
# 获取聚类标签
cluster_labels = kmeans.labels_
df['Cluster'] = cluster_labels
df.head()
```

```python
df['Cluster'].value_counts()
plt.figure(figsize=(8, 4))
sns.countplot(x='Cluster', data=df)
plt.title('Kmeans 聚类分布情况')
plt.xlabel('聚类标签')
plt.ylabel('人数')
# 显示图表
plt.show()
```

```python
df.shape
```

```python
# 获取聚类中心
cluster_centers = kmeans.cluster_centers_
cluster_centers.shape
```

```python
# 计算特征的重要性：每个特征在所有簇中心的方差
feature_variances = np.var(cluster_centers, axis=0)

# 将特征及其方差放入DataFrame，并按方差降序排列
feature_importance = pd.DataFrame({
    'Feature': df_scaled.columns,
    'Variance': feature_variances
})
feature_importance = feature_importance[feature_importance['Feature'] != 'Cluster']
feature_importance = feature_importance.sort_values(by='Variance', ascending=False)
feature_importance
```

```python
# 计算每个聚类中的性别分布
gender_counts = pd.crosstab(df['Cluster'], df['Gender'])
gender_proportions = pd.crosstab(df['Cluster'], df['Gender'], normalize='index')
# normalize='index' 参数，这意味着结果将显示每个聚类中不同性别的比例，而不是数量。
```

```python
gender_counts
```

```python
gender_proportions
```

```python
# 创建图形
plt.figure(figsize=(12, 8))

# 绘制Female部分
plt.bar(gender_counts.index, gender_counts['Female'], 
        label='Female', color='#ff9999')

# 绘制Male部分
plt.bar(gender_counts.index, gender_counts['Male'], 
        bottom=gender_counts['Female'],
        label='Male', color='#66b3ff')

for i in range(len(gender_counts)):
    # 遍历每个性别
    for gender in ['Female', 'Male']:
        proportion = gender_proportions.iloc[i][gender]
        count = gender_counts.iloc[i][gender]
        if proportion > 0.01:# 避免为0的类型也添加标签
            # 计算标签位置
            if gender == 'Female':
                y_position = count/2
            else:
                y_position = gender_counts['Female'].iloc[i] + count/2
            
            # 添加百分比标签
            plt.text(i, y_position, f'{proportion:.1%}', 
                    ha='center', va='center',
                    fontsize=10, color='white', fontweight='bold')

plt.title('各聚类的性别分布')
plt.xlabel('聚类')
plt.ylabel('人数')
plt.legend(title='性别')
plt.xticks(range(len(gender_counts)), [f'类别{i}' for i in range(len(gender_counts))])

plt.tight_layout()
plt.show()
```

```python
# 对于锻炼类型也可以用类似的代码
workout_counts = pd.crosstab(df['Cluster'], df['Workout_Type'])
workout_proportions = pd.crosstab(df['Cluster'], df['Workout_Type'], normalize='index')

plt.figure(figsize=(12, 6))

# 使用不同的颜色
colors = ['#3498db', '#e74c3c', '#2ecc71', '#f1c40f']
bottom = np.zeros(len(workout_counts))

# 绘制每种锻炼类型
for i, workout in enumerate(workout_counts.columns):
   plt.bar(workout_counts.index, workout_counts[workout],
           bottom=bottom, label=workout, color=colors[i])
   
   for j in range(len(workout_counts)):
       proportion = workout_proportions.iloc[j][workout]
       if proportion > 0.01: # 避免为0的类型也添加标签
           y_position = bottom[j] + workout_counts.iloc[j][workout]/2
           plt.text(j, y_position, f'{proportion:.1%}',
                   ha='center', va='center',
                   fontsize=10, color='white', fontweight='bold')
   # 堆叠
   bottom += workout_counts[workout]

plt.title('各聚类的锻炼类型分布')
plt.xlabel('聚类')
plt.ylabel('人数')
plt.legend(title='锻炼类型')
plt.xticks(range(len(workout_counts)), [f'类别{i}' for i in range(len(workout_counts))])

plt.tight_layout()
plt.show()
```
