## 假期出行数据全国地图分布 

```python
df_tmp1 = df[['城市','销量']]
df_counts = df_tmp1.groupby('城市').sum()
```

```python
df_counts
```

```python
df_counts.index.values
```

```python
provinces = (df_counts.index.values+'省').tolist()
sales = df_counts.values.tolist()
```

```python
# 定义颜色列表，用于视觉映射
colors = ["#D7DA8B", "#E15457", "#5C1E0F", "#6B2815", "#F3A925"]
m1 =(
    Map()
    .add('假期出行分布',[list(z) for z in zip(provinces,sales)],# 一定要加上省这个字
         'china'
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='假期出行数据地图分布'),
        visualmap_opts=opts.VisualMapOpts(max_= df_counts['销量'].max()+1, is_piecewise=False,)
    )
)
m1.load_javascript()
```

```python
m1.render_notebook()
```

```python
max_value = df_counts['销量'].max()  # 获取最大销量值
m1 = (
    Map()
    .add("假期出行分布", [list(z) for z in zip(df_counts.index.values.tolist(), df_counts['销量'].values.tolist())], "china")
    .set_global_opts(
        title_opts=opts.TitleOpts(title='假期出行数据地图分布'),
        visualmap_opts=opts.VisualMapOpts(max_=max_value, is_piecewise=False, range_color=colors)
    )
)
m1.render_notebook()
```

## 各省市4A-5A景区数量柱状图 


```python
df[df['星级'].isin(['4A', '5A'])]
```

```python
df_tmp2 = df[df['星级'].isin(['4A', '5A'])]
df_counts = df_tmp2.groupby('城市').count()['星级']
df_counts
```

```python
b2=(
    Bar()
    .add_xaxis(df_counts.index.values.tolist())
    .add_yaxis('4A-5A景区数量', df_counts.values.tolist())
    .set_global_opts(
        title_opts=opts.TitleOpts(title='各省市4A-5A景区数量'),
        datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_='inside')],
    )
)
b2.load_javascript()
```

```python
b2.render_notebook()
```

```python
# 定义 JavaScript 代码来动态设置颜色
color_js = """
function(params) {
    var colorList = ['#5470C6', '#91CC75', '#EE6666', '#73C0DE', '#3BA272', '#FC846B', '#9A60B4', '#EA7CCC'];
    return colorList[params.dataIndex % colorList.length];
}
"""

# 创建柱状图
b2 = (
    Bar()
    .add_xaxis(df_counts.index.values.tolist())
    .add_yaxis('4A-5A景区数量', df_counts.values.tolist(), itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_js)))
    .set_global_opts(
        title_opts=opts.TitleOpts(title='各省市4A-5A景区数量'),
        # 用于设置图表的缩放组件。内部缩放工具（Inside） 外部缩放工具（Slider）
        datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_='inside')],
    )
)

# 在 Jupyter Notebook 中渲染图表
b2.render_notebook()
```

## 门票价格区间占比玫瑰图 

```python
df['价格'].info()
```

```python
df['价格'].astype(float).plot()
```

```python
# 存在异常值
df['价格']  = pd.to_numeric(df['价格'], errors='coerce')
df['价格'].plot()
```

```python
price_level = [0, 50, 100, 150, 200, 250, 300, 350, 400, 500] 
label_level = ['0-50', '50-100', '100-150', '150-200', '200-250', '250-300', '300-350', '350-400', '400-500'] 
jzmj_cut = pd.cut(df['价格'], price_level, labels=label_level) 
jzmj_cut
```

```python
df_price = jzmj_cut.value_counts()
p1 = (
    Pie(init_opts=opts.InitOpts(width='800px', height='600px',)
       )
    .add(
        '',
        [list(z) for z in zip(df_price.index.tolist(), df_price.values.tolist())],
        radius=['20%', '60%'],
        center=['40%', '50%'],
        rosetype='radius',
        label_opts=opts.LabelOpts(is_show=True),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='门票价格占比玫瑰图',pos_left='33%',pos_top="5%"),
        legend_opts=opts.LegendOpts(type_='scroll', pos_left="80%",pos_top="25%",orient="vertical")
    )
)
p1.render_notebook()
```

##  景点词云图

```python
text = str(df['简介'].tolist())
# 使用 jieba 进行中文分词
words = jieba.cut(text)
word_counts = Counter(words)  # 计算词频
```

```python
word_counts
```

```python
w2 = (
    WordCloud()
    .add(
        "",
        word_counts.most_common(200),
        word_size_range=[15, 50],
        word_gap = 20,
        textstyle_opts=opts.TextStyleOpts(font_family="cursive"),# 字体设置
        shape = 'star',
        # mask_image='1.jpg',
)
    .set_global_opts(
        title_opts=opts.TitleOpts(title=" 词云图"),
         # 提示框设置
        tooltip_opts=opts.TooltipOpts(is_show=True)
    )
)
w2.load_javascript()
```

```python
w2.render_notebook()
```
